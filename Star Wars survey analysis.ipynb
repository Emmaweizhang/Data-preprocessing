### 1.4 Deal with missing values
First, check whether there are any Null existing in the dataset.
It appears that there are a large number of missing values existing in this dataset. 'ID' and 'seen_SW' columns contain no missing values but all other features have lots of missing values. By scanning through the original dataset, it's obvious to see that there are rows having null for all features. We know that there are 38 columns in total. Let's find out how many rows have 36 empty cells.

```python
starwars.isnull().sum()
```

```python
empty = np.where(starwars.isnull().sum(axis=1) == 36)
len(empty[0])
```

Therefore, there are 110 rows containing all empty cells except for 'ID' and 'seen_SW' columns. Those are respondents who only answered the first question. They have no usefull information for our analysis thus should be dropped. The new dataset `starwars1` are without the 110 rows. One might wonder, if those who have never seen star war movies left the questions blank? 
Then we count the values in 'seen_SW' column and realised that this is not the case.

```python
starwars1 = starwars.drop(starwars.index[empty])
```

```python
starwars['seen_SW'].value_counts()
```

```python
starwars1['seen_SW'].value_counts()
```

After dropping the 110 rows, we are left with 1075 rows. Among them, 240 respondents have never seen Star Wars movies. 
For those who have never seen Star Wars movies, it'd be hard to rank the movies or rate the characters. If those rows are dropped, will we end up with a dataset with much less null? The new dataset `starwars2` contains only information about those who have seen the movies.

```python
starwars2 = starwars1[starwars1.seen_SW != 'No']
starwars2.isnull().sum()
```

This is much better! The deleted 240 rows are those having missing values at ranking columns. It seems that there are lots of Null for `'seen_1' - 'seen_6'` columns. However, these 6 columns are answering the question
'Which of the following Star Wars ﬁlms have you seen?' Let's check the unique values of these features.

```python
for col in ['seen_1', 'seen_2','seen_3','seen_4','seen_5','seen_6']:
    print('The unique values of column', col, 'are')
    print(starwars2[col].unique())
    print('\n')
```

Based on the output above, each of the 6 features only has one value and null. If the column value is null, it only means that
the respondent has not seen this particular star war movie. Therefore, let's replace the values with 'Yes' and 'No'. 'Yes' means the respondent has seen this film and 'No' means the respondent has not seen it.

```python
for col in ['seen_1', 'seen_2','seen_3','seen_4','seen_5','seen_6']:
    starwars2[col].fillna('No', inplace = True) 
    starwars2.loc[starwars2[col] != 'No', col] = 'Yes'
```

```python
starwars2.isnull().sum()
```

### 1.5 Drop unuseful columns

Column 'ID' are only numbers assigned by the survey and it has no useful information for the analysis. Column 'fan_expanded' answers
question 'Do you consider yourself to be a fan of the Expanded Universe?' but the majority of respondents did not answer this question.
We do not have enough data to understand this column. Therefore, `starwars3` has 'ID' and 'fan_expanded' columns dropped.

```python
starwars3 = starwars2.drop(['ID', 'fan_expanded'], axis = 1)
```

### 1.7 Correct typos and sanity check

As per previous outputs, it's clear to see the typos in some of the columns. Let's correct them here.

```python
starwars3['fan'] = starwars3['fan'].str.replace('Yess', 'Yes')
starwars3['fan'] = starwars3['fan'].str.replace('Noo', 'No')
# Check again
starwars3['fan'].unique()
```

```python
starwars3['fan_st'] = starwars3['fan_st'].str.replace('yes', 'Yes')
starwars3['fan_st'] = starwars3['fan_st'].str.replace('Noo', 'No')
# Check again
starwars3['fan_st'].unique()
```

```python
starwars3['gender'] = starwars3['gender'].str.replace('Female', 'F')
starwars3['gender'] = starwars3['gender'].str.replace('female', 'F')
starwars3['gender'] = starwars3['gender'].str.replace('Male', 'M')
# Check again
starwars3['gender'].unique()
```

```python
# make sure '500' is not a category
m = starwars3['age'] == '500'
m.value_counts()
```

```python
#starwars3['age'] = starwars3['age'].replace('500', np.nan)
# Check again
#starwars3['age'].unique()
```

### 1.9 Deal with missing values

Missing values are still existing in the characters columns. As the percentages of missing values are small, those rows can be dropped. However, `income` column has a higher percentage of missing values than other columns. I will impute it with mode.

```python
starwars3.isnull().sum()
```

```python
for col in ['rank_1', 'rank_3','Han_S', 'Luke_S', 'Prin_L', 'Anakin_S', 'Obi_W','Emperor_P', 'Darth_V', 'Lando_C', 'Boba_F', 'C3PO', 'R2D2','Jar_J', 'Padme_A', 'Yoda', 'gender', 'age', 'edu', 'loca']:
    starwars3 = starwars3[starwars3[col].notna()]
```

```python
mode = starwars3['income'].mode()[0]
starwars3['income'].fillna(mode, inplace=True)
```
